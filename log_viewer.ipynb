```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
log2 = pd.read_json('results/baseline/log')
log = pd.read_json('results/trial/log')
log_comparision = log.merge(log2, left_on='epoch', right_on='epoch', suffixes=('/baseline', '/trial'), how='outer')
log_comparision[:].plot(x='epoch', y=['main/accuracy/baseline', 'validation/main/accuracy/baseline', 'main/accuracy/trial', 'validation/main/accuracy/trial'])
log_comparision[:].plot(x='epoch', y=['main/loss/baseline', 'validation/main/loss/baseline', 'main/loss/trial', 'validation/main/loss/trial'])

display(log_comparision)
```
